```python
import numpy as np

# Определение функции
def function(x, y):
    return x**3 + 4*x*y**2 + 32*y**2 + 9*x*y + 10*y + 20*x**2 + 22*x + y**3 + 97

# Градиент функции
def gradient(x, y):
    dx = 3*x**2 + 4*y**2 + 9*y + 40*x + 22
    dy = 8*x*y + 64*y + 9*x + 3*y**2 + 10
    return np.array([dx, dy])

# Градиентный спуск
def gradient_descent(x, y, learning_rate, epsilon, max_iterations):
    iteration = 0
    while True:
        grad = gradient(x, y)
        x_new = x - learning_rate * grad[0]
        y_new = y - learning_rate * grad[1]
        if np.linalg.norm([x_new - x, y_new - y]) < epsilon or iteration > max_iterations:
            break
        x, y = x_new, y_new
        iteration += 1
    return x_new, y_new

# Стохастический спуск
def stochastic_gradient_descent(x, y, learning_rate, epsilon, max_iterations):
    iteration = 0
    while True:
        x_new = x - learning_rate * (3*x**2 + 4*y**2 + 9*y + 40*x + 22)
        y_new = y - learning_rate * (8*x*y + 64*y + 9*x + 3*y**2 + 10)
        if np.linalg.norm([x_new - x, y_new - y]) < epsilon or iteration > max_iterations:
            break
        x, y = x_new, y_new
        iteration += 1
    return x_new, y_new
```

```python
# Определение начальной точки, шага и точки останова
x = 0
y = 0
learning_rate = 0.01
epsilon = 1e-6
max_iterations = 1000

# Вызов функций градиентного и стохастического спуска
result_gradient_descent = gradient_descent(x, y, learning_rate, epsilon, max_iterations)
result_stochastic_gradient_descent = stochastic_gradient_descent(x, y, learning_rate, epsilon, max_iterations)

print("Минимум функции (градиентный спуск): x = {:.4f}, y = {:.4f}".format(result_gradient_descent[0], result_gradient_descent[1]))
print("Минимум функции (стохастический спуск): x = {:.4f}, y = {:.4f}".format(result_stochastic_gradient_descent[0], result_stochastic_gradient_descent[1]))
```
